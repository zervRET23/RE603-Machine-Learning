**Function Parameter dengan Arbitrary Arguments (Hands On )**




Buatlah fungsi untuk mencari "***Luas Lingkaran dan Keliling Lingkaran***" dengan jari-jari 15 cm.

1. Luas Lingkaran = 3.14 * jari-jari**2
2. keliling Lingkaran = 2 * 3.14 * jari-jari

```python
# Function untuk menghitung luas lingkaran
def luas_lingkaran(radius):
    pi = 3.14
    Luas = pi * radius** 2
    return Luas
# Function untuk menghitung keliling lingkaran
def kel_lingkaran(radius):
    pi = 3.14
    Keliling = 2 * pi * radius
    return Keliling

print("Luas Lingkaran:", luas_lingkaran(15))
print("Keliling Lingkaran:", kel_lingkaran(15))
```

Buat suatu kelas bernama User


```python
# Membuat kelas User
class User:
    def __init__(self, username, email, umur):
        self.username = username
        self.email = email
        self.umur = umur

    def info_user(self):
        return f"Username: {self.username}, Email: {self.email}, Umur: {self.umur}"

# Membuat objek dari kelas User
user_obj = User("Zerv", "zervret@gmail.com", "20")

# Menampilkan informasi pengguna
print(f"""
--- Informasi Pengguna ---
Username: {user_obj.username}
Email: {user_obj.email}
Umur: {user_obj.umur}
"""
)
```

**Instance Variable - OOP (Hands On )**




Buatlah class Lingkaran yang memiliki atribut : jari-jari 10 cm, pi = 3.14159
dengan method Luas Lingkaran dan Keliling Lingkaran

```python
class Lingkaran:
  def __init__(self):
    self.jari_jari = 10
    self.pi = 3.14159

  def luas_lingkaran(self):
    return self.pi * self.jari_jari**2

  def keliling_lingkaran(self):
    return 2 * self.pi * self.jari_jari

# Tampilkan hasil
lingkaran = Lingkaran()
print(f"Jari-jari: {lingkaran.jari_jari} cm")
print(f"Luas Lingkaran: {lingkaran.luas_lingkaran():.2f} cm²")
print(f"Keliling Lingkaran: {lingkaran.keliling_lingkaran():.2f} cm")
```
